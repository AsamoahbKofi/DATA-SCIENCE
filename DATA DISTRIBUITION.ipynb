SKEW RIGHT DATA

Data with a skew to the right, also called positive skew, has some interesting characteristics that

can be helpful in understanding the underlying phenomenon it represents.'

Here are some deductions you can draw from a right-skewed data set:

Presence of outliers: A right skew suggests there are more data points on the left side of the distribution compared to the right.

This often indicates the presence of outliers, which are extreme values that fall far away from the majority of the data


Long tail: The tail of the distribution on the right side stretches out further compared to a normal distribution.

This "long tail" signifies a higher frequency of values in the lower range and a smaller but impactful presence of much higher values.

Mean greater than median: In a right-skewed distribution, the mean (average) is usually located further to the right compared to the median (middle value)
.
 This is because the mean is pulled towards the tail by the influence of the larger values.

Basically, majority of the data points are around the mean with few points very high

```python
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# Generate skewed right data
np.random.seed(0)
data = np.random.exponential(scale=1, size=1000)

# Plot the skew right data using Seaborn
sns.histplot(data, kde=True, color='skyblue', stat='density')
sns.set_style("whitegrid")
sns.despine(left=True)
```

SKEW LEFT DATA

Predominance of lower values: A left skew indicates that most of the data points 

are concentrated on the right side of the distribution, compared to the left side. 

This suggests a prevalence of lower values in the data set.
    
Short tail on the left: The tail of the distribution on the left side is shorter 

compared to a normal distribution. This signifies a smaller number of data points in the

lower range and a concentration of values in the higher range.

Median greater than mean: In a left-skewed distribution, the median (middle value)

is usually located further to the right compared to the mean (average). 

This is because the mean is pulled towards the lower tail by the influence of the smaller values.


```python
import seaborn as sns
import numpy as np

# Generate skewed left data
np.random.seed(0)
data = np.random.chisquare(df=1, size=1000)

# Plot the skew left data using Seaborn
sns.histplot(data, kde=True, color='salmon', stat='density')
sns.set_style("whitegrid")
sns.despine(left=True)
```

NORMAL DISTRIBUTION

Mean and Median: In a normal distribution, the mean, median,

and mode are all equal and located at the center of the distribution.

Standard Deviation: The spread of the data in a normal distribution is
determined by the standard deviation. 
    
About 68% of the data falls within one standard deviation of the mean (±1σ),

approximately 95% within two standard deviations (±2σ), and nearly all (99.7%)

within three standard deviations (±3σ) from the mean.

Empirical Rule: The Empirical Rule states that in a normal distribution, 

approximately 68% of the data falls within one standard deviation of the mean,

95% within two standard deviations, and 99.7% within three standard deviations.

Z-Scores: Z-scores, also known as standard scores,

indicate how many standard deviations a data point is from the mean of the distribution. 

They are calculated as the difference between the observed value and the mean, divided by the standard deviation.

Understanding the Central Limit Theorem


The Idea:
The Central Limit Theorem (CLT) states that the sampling distribution

of the sample mean approaches a normal distribution as the sample size increases,

regardless of the shape of the population distribution.
    
In simpler terms, it tells us that if we take multiple samples from any population 

and calculate the mean of each sample, then those sample means will be normally distributed,

regardless of the shape of the original population.

Key Points:

Sample Size Matters: The CLT holds true particularly when the sample size is sufficiently large 

(usually considered to be at least 30 observations).

Independence: The samples must be independent of each other.
    
Random Sampling: Samples must be selected randomly from the population.

Suppose we're interested in the average height of people in a city.

We take multiple random samples of different sizes

(e.g., 10 people, 100 people, 1000 people) and calculate the mean height for each sample.
                                                                        
According to the CLT, as we increase the sample size, the distribution of these sample means

will approach a normal distribution, regardless of the shape of the original population's

height distribution. This means that even if the heights in the city are not normally distributed, 

the distribution of sample means will be.

LIMITATIONS

The CLT may not apply if the samples are not independent

or if the population distribution has very heavy tails or extreme skewness.
